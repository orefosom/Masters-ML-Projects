Task (a)

Make a plot to visualize industry vs. academia researchers distribution for each year from 2023 up to 2123 according to our model. Explain what you see.

```python

import numpy as np
import matplotlib.pyplot as plt

# Define the transition matrix T
T = np.array([[0.99, 0.2], [0.01, 0.8]])

# Define the initial distribution in 2023
x0 = np.array([0.3, 0.7])

# Initialize a list to store the distributions over time
distributions = [x0]

# Calculate the distributions for the next 100 years
for _ in range(100):
    x0 = np.dot(T, x0)
    distributions.append(x0)

# Convert the list of distributions into a NumPy array
distribution_array = np.array(distributions)

# Create a plot to visualize the distribution over time
years = range(2023, 2124)
plt.figure(figsize=(10, 6))
plt.plot(years, distribution_array[:, 0], label='Industry', marker='o')
plt.plot(years, distribution_array[:, 1], label='Academia', marker='o')
plt.xlabel('Year')
plt.ylabel('Distribution')
plt.title('Industry vs. Academia Researchers Distribution Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

Observations:

(1) As time progresses, the distribution will approach a steady-state or equilibrium.

(2) The transition matrix T favors stability in the long run, with researchers mostly staying in their current sector. However, initially, there is a transition, but over time, the distribution stabilizes, with a majority of researchers remaining in their respective sectors.
