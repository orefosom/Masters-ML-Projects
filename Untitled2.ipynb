```python

# Define the prior probability (initial belief) of getting heads
prior_probability_heads = 0.5

# Define the likelihood of getting heads (in this case, 0.5 since it's a fair coin)
likelihood_heads = 0.5

# Perform 7 consecutive tosses with all heads
num_tosses = 7
for _ in range(num_tosses):
    # Calculate the posterior probability using Bayes' theorem
    posterior_probability_heads = (prior_probability_heads * likelihood_heads) / ((prior_probability_heads * likelihood_heads) + ((1 - prior_probability_heads) * (1 - likelihood_heads)))

    # Update the prior probability for the next toss
    prior_probability_heads = posterior_probability_heads

print(f"The posterior probability of getting heads after 7 consecutive heads is: {posterior_probability_heads:.4f}")
```

```python

# Define the prior probability of getting heads (initial belief that the coin is fair)
prior_probability = 0.5

# Define the number of tosses
num_tosses = 7

# List to store the posterior probabilities after each toss
posterior_probabilities = []

for _ in range(num_tosses):
    # Likelihood: Probability of getting heads (assuming the coin is fair, it's 0.5)
    likelihood = 0.5

    # Update the posterior probability using Bayes' theorem
    posterior_probability = (prior_probability * likelihood) / (prior_probability * likelihood + (1 - prior_probability) * (1 - likelihood))

    # Update the prior probability for the next toss
    prior_probability = posterior_probability

    # Append the posterior probability to the list
    posterior_probabilities.append(posterior_probability)

# The final posterior probability after 7 consecutive heads is the result
final_posterior_probability = posterior_probabilities[-1]

print("Final posterior probability after 7 consecutive heads:", final_posterior_probability)
```

```python

def display_penguins(n):
    penguin = [
        "   _~_   ",
        "  (o o)  ",
        " /  V  \\ ",
        "/(  _  )\\",
        "  ^^ ^^  "
    ]
    for i in range(len(penguin)):
        for j in range(n):
            print(penguin[i], end=" ")
        print()


if __name__ == "__main__":
    try:
        N = int(input("Enter a natural number (1 to 9): "))
        if 1 <= N <= 9:
            display_penguins(N)
        else:
            print("Please enter a natural number between 1 and 9.")
    except ValueError:
        print("Invalid input. Please enter a valid natural number.")
```

```python

import numpy as np
import sympy

# Given matrix A
A = np.array([[5, -6, -6],
              [-1, 4, 2],
              [3, -6, -4]])

# Step 1: Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Step 2: Form matrix P with eigenvectors as columns
P = eigenvectors

# Step 3: Calculate the inverse of P
P_inv = np.linalg.inv(P)

# Step 4: Diagonal matrix D
D = np.diag(eigenvalues)

# Step 5: Verify A = P * D * P_inv
reconstructed_A = P @ D @ P_inv

Alternate = sympy.Matrix(A).diagonalize()

# Print results
print("Original Matrix A:")
print(A)
print("\nEigenvalues:")
print(eigenvalues)
print("\nEigenvectors (as columns):")
print(eigenvectors)
print("\nMatrix P:")
print(P)
print("\nMatrix P inverse:")
print(P_inv)
print("\nDiagonal Matrix D:")
print(D)
print("\nReconstructed A (should be equal to original A):")
print(reconstructed_A)
print("\nDiagonalize using sympy library")
print(Alternate)
```

```python

from sympy import Matrix

# Define your 3x3 matrix as a Sympy Matrix
matrix = Matrix([[5, -6, -6],
                 [-1, 4, 2],
                 [3, -6, -4]])

# Find eigenvalues and eigenvectors
eigeninfo = matrix.eigenvects()

# Extract eigenvalues and eigenvectors from the result
eigenvalues = [ei[0] for ei in eigeninfo]
eigenvectors = [ei[2][0] for ei in eigeninfo]

# Print the results
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:")
for i, vector in enumerate(eigenvectors):
    print(f"Eigenvalue {i + 1}: {vector}")
```
