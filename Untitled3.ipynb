```python

import types
from collections.abc import Container

def hacker_decorator(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        if isinstance(result, (int, float)):
            return result * 2
        elif isinstance(result, Container):
            return type(result)()
        return result
    return wrapper

def apply_hacker_decorator(cls):
    for name, method in vars(cls).items():
        if isinstance(method, types.FunctionType) and not name.startswith('_'):
            setattr(cls, name, hacker_decorator(method))
    return cls

class HackedClass:
    pass  # Add your original class definition here

# Apply the hacker_decorator to HackedClass
HackedClass = apply_hacker_decorator(HackedClass)
```
